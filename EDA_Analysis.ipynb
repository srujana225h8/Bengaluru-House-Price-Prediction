```python
# import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

**Exploratory Data Analysis**

```python
# Dropping unwanted columns
cols = ["area_type", "availability", "society", "balcony"]
df = df.drop(cols, axis = 1)
df.head()
```

```python
df.info()
```

```python
# null values
df.isnull().sum()
```

```python
# dropping duplicates if present
df.drop_duplicates(inplace=True)
```

**Handling null values**

```python
# cleaning location columns

# removing trailing and leading spaces for location
df['location'] = df['location'].str.strip()

# handling NaN values in location with top 2nd location
df['location'] = df['location'].fillna("Sarjapur  Road")
df.isnull().sum()

# grouping the locations with less than 10 houses into 'others' category

loc = df['location'].value_counts()
loc_less_than_10 = loc[loc<=10]
df['location'] = df['location'].apply(lambda x : "others" if x in loc_less_than_10 else x)
```

```python
# cleaning size column

# handling NaN values
df['size'] = df['size'].fillna("2 BHK")

# extract only the BHK number and remove unwanted values
df['size'] = df['size'].apply(lambda x: int(x.split()[0]))
df['size'].unique()

# renaming size to BHK
df.rename(columns={"size":"BHK"}, inplace=True)
df.head()
```

```python
# cleaning Total square feet column

# splitting the range type values
def clean_total_sqft(sqft):
    try:
        l = sqft.split("-")
        if len(l)==2:
            return (float(l[0])+float(l[1]))/2
        else:
            return float(l[0])
    except:
        return None
    
df['total_sqft'] = df['total_sqft'].apply(clean_total_sqft)

# handling the null values
df['total_sqft'] = df['total_sqft'].fillna(round(df["total_sqft"].mean()))
```

```python
# cleaning the bath column

# handling null values in bath column
med = df["bath"].median()
df['bath'] = df["bath"].fillna(med)
df["bath"].unique()
```

**Handling Outliers**

```python
# create a new col price_per_sqft

# formula: price_per_sqft = price/total_sqft
df["price_per_sqft"] = (df['price']*100000)/df["total_sqft"]
df
```

```python
# handling outliers in total_sqft

# observation: In total_sqft we have houses less than 300 sqft, those are outliers
# In real-time houses starts with minimum of 300 sqft
df.describe()

# formula: total_sqft/BHK >= 300 ---> It keeps all the realistic total sqft

df = df[df["total_sqft"]/df["BHK"]>=300]
```

```python
# all unrealistic rows with total_sqft less than or equal to 300 are removed
df.describe()
```

```python
# removing the outliers in BHK column
df["BHK"].unique()
df = df[df["BHK"]<=6]
sns.boxplot(x="BHK", data = df)
```

```python
# Identifying outliers in bath colummn
df[df["bath"]>df["BHK"]+2][["BHK","bath"]]

# Removing the outliers
df = df[df["bath"]<df["bath"]+2]
```

```python
# identifying the outliers in price_per_sqft column and dropping them

sns.boxplot(x="price_per_sqft",data = df)

q1 = df["price_per_sqft"].quantile(0.25)
q3 = df["price_per_sqft"].quantile(0.75)
IQR = q3 - q1
lower = q1-0.5*IQR
upper = q3+0.5*IQR
df = df[(df['price_per_sqft']>=lower) & (df["price_per_sqft"]<=upper)]
```

```python
sns.boxplot(x="price_per_sqft",data = df)
```

```python
# remove price_per_sqft 
df = df.drop('price_per_sqft', axis = 1)
```

```python
df.reset_index(inplace=True)
df = df.drop("index", axis = 1)
df
```

```python
data = df.groupby("location")["price"].mean().sort_values()
plt.bar(data.index, data)
plt.xticks(rotation= 90)
plt.show()
```

**Model Training & Evaluation**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#Encoding Location column
encoder = LabelEncoder()
df["encoded_loc"] = encoder.fit_transform(df["location"])
```

```python
#Divide data into input and output features
X = df.drop(["location","price"],axis=1)
y = df.price
```

```python
#split the data for training and testing
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#model training
LR = LinearRegression()
LR.fit(Xtrain,ytrain)
```

```python
ypred = LR.predict(Xtest)
ypred
```

```python
print("Training Eff: ",LR.score(Xtrain,ytrain))
print("Testing Eff: ",LR.score(Xtest,ytest))
print("R2: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

```python
DT = DecisionTreeRegressor()
DT.fit(Xtrain,ytrain)
```

```python
ypred = DT.predict(Xtest)
ypred
```

```python
print("Training Eff: ",DT.score(Xtrain,ytrain))
print("Testing Eff: ",DT.score(Xtest,ytest))
print("R2: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

```python
#using pickle
#Model serialization-->Saving the trained model
#req:  it skips the continuous training process
import pickle
with open("Linear_model.pkl","wb") as file:
    pickle.dump(LR,file)
```

```python
#encoder serialization
with open("label_encoder.pkl","wb") as file:
    pickle.dump(encoder,file)
```
